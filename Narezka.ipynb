```python
import cv2, numpy as np, matplotlib
from matplotlib import pyplot as plt

image = cv2.imread('Pek.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
h,w = image.shape
image = image[100:-200,130:1010]

ret,thresh = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

kernel = np.ones((5,25), np.uint8)
img_dilation = cv2.dilate(thresh, kernel, iterations=1)

roi = img_dilation[200:700]

plt.figure(figsize=(20,30))
plt.imshow(roi, cmap='gray')
```

```python
sums = np.sum(roi, axis=0)
plt.figure(figsize=(25,5))
plt.plot(sums[200:700])
```
